```python
import pandas as pd
data=pd.read_csv('data2.csv')

X=data['gdp']
Y=data['h_score']

import numpy as np

degri=15
fit=np.polyfit(X,Y,degri)

```

```python
import matplotlib.pyplot as plt

%pylab inline

plt.scatter(X,Y)
pred=0
for i in xrange(degri+1):
    pred+=(X**i)*fit[degri-i]

x,y=zip(*(sorted(zip(X,pred))))
plt.plot(x,y,color='red')
```
